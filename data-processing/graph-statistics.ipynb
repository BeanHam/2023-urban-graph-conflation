```python
import os
import numpy as np
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
from tqdm import tqdm
```

## Sanity Check

```python
def statistics(data_path, source_name):
    path = data_path+f"{source_name}/"
    files = os.listdir(path)
    files = [f for f in files if 'DS_Store' not in f]
    node_length = []
    edge_length = []    
    for file in tqdm(files):
        G = nx.read_graph6(path+file)
        node_length.append(len(G.nodes()))
        edge_length.append(len(G.edges())) 
    return node_length, edge_length
```

```python
## load seattle city blocks
data_path = '../../../data/2023-graph-conflation/graphs/'
osw_node, osw_edge = statistics(data_path, 'osw')
osm_node, osm_edge = statistics(data_path, 'osm')
sdot_node, sdot_edge = statistics(data_path, 'sdot')
```

```python
plt.hist(osw_node, bins=50, alpha=0.4, label='OSW')
plt.hist(osm_node, bins=50, alpha=0.4, label='OSM')
plt.hist(sdot_node, bins=50, alpha=0.4, label='SDOT')
plt.title('Node Count Distribution', fontsize=20)
plt.legend()
plt.text(100, 1000, f"Min: {np.min(osw_node)}")
plt.text(100, 900, f"25%: {np.quantile(osw_node, 0.25)}")
plt.text(100, 800, f"50%: {np.quantile(osw_node, 0.5)}")
plt.text(100, 700, f"75%: {np.quantile(osw_node, 0.75)}")
plt.text(100, 600, f"95%: {np.quantile(osw_node, 0.95)}")
plt.text(100, 500, f"Max: {np.max(osw_node)}")
pass
```
