```python
import osmnx as ox
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from tqdm import tqdm
```

```python
def post_processing(pois):
    
    ## extract each feature vector
    try: bus_stop = pois.highway.values
    except: bus_stop = 0
        
    try: crossing = pois.crossing.values
    except: crossing = 0
    
    try: leisure = pois.leisure.values
    except: leisure = 0
                
    try: building = pois.building.values
    except: building = 0
    
    try: shop = pois.shop.values
    except: shop = 0
    
    try: amenity = pois.amenity.values
    except: 
        amenity = 0
        food = 0
        education = 0
        financial = 0
        healthcare = 0
        
    ## count
    if type(bus_stop) == np.ndarray: bus_stop = len(bus_stop[bus_stop == 'bus_stop'])
    if type(crossing) == np.ndarray: crossing = sum(~pd.isnull(crossing))
    if type(leisure) == np.ndarray: leisure = sum(~pd.isnull(leisure))
    if type(building) == np.ndarray: building = sum(~pd.isnull(building))
    if type(shop) == np.ndarray: shop = sum(~pd.isnull(shop))
    if type(amenity) == np.ndarray:
        amenity = amenity[~pd.isnull(amenity)]    
        food = len(amenity[np.isin(amenity, ["restaurant", "bar", "cafe", "fast_food", "pub"])])
        education = len(amenity[np.isin(amenity, ["school", "college", "kindergarten", "library", "university"])])
        financial = len(amenity[np.isin(amenity, ["bank", "atm"])])
        healthcare = len(amenity[np.isin(amenity, ["hospital", "clinic", "dentist", "doctors", "pharmacy"])])
    
    return [crossing, bus_stop,leisure, building, shop, food, education, financial, healthcare]
```

```python
tags = {
    ## extract crossings
    "crossing": True,
        
    ## features
    "highway": "bus_stop", #Pedestrians can cross a street here; e.g., zebra crossing
    
    "leisure": [
        "park", "garden", "fitness_station", 
        "fitness_centre", "playground", "sports_centre", 
        "stadium"
    ],
    "amenity": [
        ## food related
        "restaurant", "bar", "cafe", "fast_food", "pub",
        
        ## education
        "school", "college", "kindergarten", "library", "university",
        
        ## financial
        "bank", "atm",
        
        ## healthcare
        "hospital", "clinic", "dentist", "doctors", "pharmacy"
    ],
    "building": [
        "apartments", "dormitory", "hotel", 
        "house", "residential", "commercial",
        "industrial", "office", "retail",
        "supermarket", "warehouse", ""
    ],
    "shop": True
    
}
```

### Area

```python
seattle = gpd.read_file('Neighborhood_Map_Atlas_Neighborhoods.shp')
seattle = seattle.to_crs(epsg=4326)
seattle.plot()
```

```python
seattle.head()
```

### Extract Features

```python
features = []
for s in tqdm(seattle.geometry):
    ## extract features for each subarea
    try:
        pois = ox.features_from_polygon(s,tags)
        count_vector = post_processing(pois)
        features.append(count_vector)
    except:
        features.append(np.random.randint(0,1,9).tolist())
```

```python
seattle_features = np.stack(features)
```

```python
np.save('seattle_features.npy', seattle_features)
```
