```python
import os
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
seattle_files = os.listdir('../graph-data/seattle-graphs/original/')
west_seattle_files = os.listdir('../graph-data/west-seattle-graphs/original/')
```

```python
dims = []
for file in seattle_files:
    G = nx.read_gpickle('../graph-data/seattle-graphs/original/'+file)
    dims.append(len(G.edges()))
```

```python
plt.hist(dims,bins=50)
pass
```
