```python
import numpy as np
import random
import geojson
import momepy
import networkx as nx
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely import LineString, MultiLineString, Polygon
from tqdm import tqdm
from random import sample
from itertools import combinations
import warnings
warnings.filterwarnings("ignore")
```

```python
def geojoin(geodata, block):
    lines = gpd.sjoin(geodata, block)
    lines = lines[['geometry']]
    G_lines = momepy.gdf_to_nx(lines, approach='primal',multigraph=False)
    return lines, G_lines
```

```python
def visualize(lines1, G_lines1, lines2, G_lines2):
    
    # plot
    f, ax = plt.subplots(2, 3, figsize=(12, 6), sharex=True, sharey=True)
    for row in ax:
        for col in row:
            col.axis("off")
    
    lines1.plot(color='#e32e00', ax=ax[0,0])
    nx.draw(G_lines1, {n:[n[0], n[1]] for n in list(G_lines1.nodes)}, ax=ax[0,1], node_size=15)
    lines1.plot(color='#e32e00', ax=ax[0,2], zorder=-1)
    nx.draw(G_lines1, {n:[n[0], n[1]] for n in list(G_lines1.nodes)}, ax=ax[0,2], node_size=15)
    
    lines2.plot(color='#e32e00', ax=ax[1,0])
    nx.draw(G_lines2, {n:[n[0], n[1]] for n in list(G_lines2.nodes)}, ax=ax[1,1], node_size=15)
    lines2.plot(color='#e32e00', ax=ax[1,2], zorder=-1)
    nx.draw(G_lines2, {n:[n[0], n[1]] for n in list(G_lines2.nodes)}, ax=ax[1,2], node_size=15)
    
    f, ax = plt.subplots(1, 1, figsize=(18, 6), sharex=True, sharey=True)
    osw_lines.plot(color='#e32e00', ax=ax, zorder=-1, label='osw')
    sdot_lines.plot(color='green', ax=ax, zorder=-1, label='sdot')
    plt.legend(loc=0)
    ax.axis("off")
```

```python
## load seattle city blocks
data_path = '../../data/2023-graph-conflation/'
seattle = gpd.read_file(data_path+'city-blocks/2010_Census_Block_Seattle_-_Housing_Statistics.shp')
west_seattle_tract = 100*np.array([96, 97.01, 97.02, 98, 99,105,106,107.01,107.02,108,112,113,114.01,114.02,115,116,120,121,264,265])
seattle = seattle.to_crs(epsg=4326)
west_seattle = seattle[np.isin(seattle.TRACT_10, west_seattle_tract)]
rest_seattle = seattle[~np.isin(seattle.TRACT_10, west_seattle_tract)]
```

```python
## visualization
f, ax = plt.subplots(1, figsize=(8,8))
ax = west_seattle.plot(ax=ax)
ax.set_title('West Seattle', fontsize=20)
ax.set_axis_off()
ax.margins(0)
ax.apply_aspect()
pass
```

```python
## visualization
f, ax = plt.subplots(1, figsize=(8,8))
ax = rest_seattle.plot(ax=ax)
ax.set_title('Main Seattle', fontsize=20)
ax.set_axis_off()
ax.margins(0)
ax.apply_aspect()
pass
```

## OSW Data

```python
## load OSW data
# footway: 139,397
# primary: 4135
# residential: 47671
# secondary: 9719
# service: 54102
# tertiary: 10508
# trunk: 225
# unclassified: 1935
path_to_file = data_path+'wa.seattle.graph.edges.Ricky.geojson'
with open(path_to_file) as f:
    gj = geojson.load(f)
osw_ori = gj['features']
```

```python
osw = [f for f in osw_ori if f['properties']['highway'] in ['footway', 'residential', 'service']]
osw = gpd.GeoDataFrame(
    [LineString(osw_ori[i]['geometry']['coordinates']) for i in range(len(osw))], 
    columns=['geometry'],
    crs=4326
)
osw.shape
```

## SDOH Data

```python
## load sdot data
sdot = gpd.read_file(data_path+'/sdot/Street_Network_Database_SND.shp')
sdot = sdot.to_crs(4326)
sdot.shape
```

```python
def segments(curve):
    return list(map(LineString, zip(curve.coords[:-1], curve.coords[1:])))
```

```python
i=100
geom = rest_seattle.iloc[[i]]
osw_lines, G_osw_lines = geojoin(osw, geom)
sdot_lines, G_sdot_lines = geojoin(sdot, geom)
```

```python
# plot
f, ax = plt.subplots(2, 3, figsize=(12, 6), sharex=True, sharey=True)
for row in ax:
    for col in row:
        col.axis("off")

osw_lines.plot(color='#e32e00', ax=ax[0,0])
nx.draw(G_osw_lines, {n:[n[0], n[1]] for n in list(G_osw_lines.nodes)}, ax=ax[0,1], node_size=15)
osw_lines.plot(color='#e32e00', ax=ax[0,2], zorder=-1)
nx.draw(G_osw_lines, {n:[n[0], n[1]] for n in list(G_osw_lines.nodes)}, ax=ax[0,2], node_size=15)

sdot_lines.plot(color='#e32e00', ax=ax[1,0])
nx.draw(G_sdot_lines, {n:[n[0], n[1]] for n in list(G_sdot_lines.nodes)}, ax=ax[1,1], node_size=15)
sdot_lines.plot(color='#e32e00', ax=ax[1,2], zorder=-1)
nx.draw(G_sdot_lines, {n:[n[0], n[1]] for n in list(G_sdot_lines.nodes)}, ax=ax[1,2], node_size=15)

f, ax = plt.subplots(1, 1, figsize=(18, 6), sharex=True, sharey=True)
osw_lines.plot(color='#e32e00', ax=ax, zorder=-1, label='osw')
sdot_lines.plot(color='green', ax=ax, zorder=-1, label='sdot')
plt.legend(loc=0)
ax.axis("off")
```

## OSM Data

```python
i=100
geom = rest_seattle.iloc[[i]]
osw_lines, G_osw_lines = geojoin(osw, geom)
sdot_lines, G_sdot_lines = geojoin(sdot, geom)
```

```python
import osmnx as ox
tags = {"highway": ["footway", "residential", "services"]}
osm_lines = ox.features_from_polygon(geom.geometry.values[0],tags)
G_osm_lines = momepy.gdf_to_nx(osm_lines, approach='primal',multigraph=False)
```

## Three Sources 
- OSW: ['footway', 'residential', 'service']
- OSM: ['footway', 'residential', 'service']
- SDOT: ['street']

```python
# plot
f, ax = plt.subplots(3, 2, figsize=(6, 6), sharex=True, sharey=True)
for row in ax:
    for col in row:
        col.axis("off")

osw_lines.plot(color='blue', ax=ax[0,0])
nx.draw(G_osw_lines, {n:[n[0], n[1]] for n in list(G_osw_lines.nodes)}, ax=ax[0,1], node_size=15)
ax[0,0].set_title('OSW')
ax[0,1].set_title('OSW Graph')

osm_lines.plot(color='red', ax=ax[1,0])
nx.draw(G_osm_lines, {n:[n[0], n[1]] for n in list(G_osm_lines.nodes)}, ax=ax[1,1], node_size=15)
ax[1,0].set_title('OSM')
ax[1,1].set_title('OSM Graph')

sdot_lines.plot(color='green', ax=ax[2,0])
nx.draw(G_sdot_lines, {n:[n[0], n[1]] for n in list(G_sdot_lines.nodes)}, ax=ax[2,1], node_size=15)
ax[2,0].set_title('SDOT')
ax[2,1].set_title('SDOT Graph')
pass
```

```python
f, ax = plt.subplots(1, 2, figsize=(6, 6), sharex=True, sharey=True)
osw_lines.plot(color='blue', ax=ax[0], zorder=-1, label='osw')
osm_lines.plot(color='red', ax=ax[0], zorder=-1, label='osm')
ax[0].set_title('OSW v.s. OSM')
ax[0].axis("off")
ax[0].legend(loc=0)

osw_lines.plot(color='blue', ax=ax[1], zorder=-1, label='osw')
sdot_lines.plot(color='green', ax=ax[1], zorder=-1, label='sdot')
ax[1].set_title('OSW v.s. SDOT')
ax[1].axis("off")
ax[1].legend(loc=0)
pass
```
